```python
import pandas as pd
import os
from glob import glob
import ast
import re
import spacy
import time

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, classification_report, adjusted_rand_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Access dataset files
folder_path = os.path.join( "..", "Data") 
file_type = "*.csv"

# List of dataset file paths
document_path = glob(os.path.join(folder_path, file_type))

document_path
```

```python
model_label = '' 
human_label = ''

for path in document_path:
    if "raw_label" in path:
        model_label = model_label + path
    elif "human_label" in path:
        human_label = human_label + path

print (model_label, human_label)
    
```

```python
#import cleaned data

def list_converter(text):
    #to revert list->str conversion from pd.read_csv
    return ast.literal_eval(text)


model_labelled_data = pd.read_csv(model_label, converters ={'tokens':list_converter})
human_labelled_data = pd.read_csv(human_label, converters ={'tokens':list_converter})
#test_data = test_data.drop(columns = ['index', 'clean_text', 'tokens', 'word_count'])
```

```python
model_labelled_data.drop(columns = ['Unnamed: 0'], inplace = True)
model_labelled_data.head()
```

```python
model_labelled_data.sort_values?
```

```python
#split `topic_description` into two columns for topic number and topic description
model_labelled_data[['topic_no', 'topics_desc']] = model_labelled_data['topic_description'].str.split('_', n = 1, expand = True)
model_labelled_data['topic_no'] = model_labelled_data['topic_no'].astype(int) 
model_labelled_data.drop(columns = ['topic_description'], inplace = True)
```

```python
#sort by `ID`
model_labelled_data.sort_values(by=['ID'], ignore_index = True, inplace = True)
model_labelled_data.head()
```

```python
model_labelled_data.info()
```

```python
#dataset showing the human labels
human_labelled_data.head()
```

```python
#include column in model_labelled_data to hold the human topic numbers

model_labelled_data['human_label'] = human_labelled_data['topic_number_bert_raw'].copy()
```

```python
topic_range = sorted(model_labelled_data['topic_no'].unique())

for topic_no in topic_range:
    mask = model_labelled_data.human_label == topic_no
    topic_df = model_labelled_data[mask]
    
    #accuracy
    y_true = topic_df.human_label
    y_pred = topic_df.topic_no
    accuracy = accuracy_score(y_true, y_pred, normalize = True)
    #class_report = classification_report(y_true,y_pred)
    #ARI = adjusted_rand_score(y_true, y_pred)
    print (f'for topic {topic_no}:Accuracy = {accuracy}')#\nClassification Report = {class_report}\nARI = {ARI}')
    
    
```
